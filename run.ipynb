```python
frames_dir = "/home/alvaro/Documents/GitHub/neural-style/examples/7fpsframes/"
out_dir = "/home/alvaro/Documents/GitHub/neural-style/examples/stylized_7fps/"
```

```python
import os
inframe_basenames = sorted(os.listdir(frames_dir))
```

### run neural style

```python
for i in inframe_basenames:
    print(i)
    content = os.path.join(frames_dir, i)
    output = os.path.join(out_dir, i)
    if(not os.path.isfile(output)):
        cmd = "python neural_style.py --content %s --styles examples/1-style.jpg --output %s --iterations 1000 --print-iterations 0" % (content, output)
        os.system(cmd)
    else:
        print("skipping")
```

### convert frames to video

```python
ffmpeg -framerate 30 -pattern_type glob -i '*.png' \
  -c:v libx264 -pix_fmt yuv420p out.mp4
```
